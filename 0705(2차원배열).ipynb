```python
def get_sum(a,b): #함수 정의(매개변수 있음, 리턴값 있음)
  s = a + b  #s = 지역변수(get_sum에서만 사용가능)
  return s
result = get_sum(5,7)
print(result)
print(get_sum(4,5))
```

```python
def sum_many(*C401): #복수 개의 가변적인 입력값이 들어올 때 * 사용
  sum = 0
  for i in C401:
    sum += i
  return sum
print(sum_many(1,2,5,41,10))
```

```python
#--------------------------------------------(정의)
def add(num1,num2):
  print(num1, '+', num2, '=', num1+num2)
def sub(num1,num2):
  print(num1, '-', num2, '=', num1-num2)
def mul(num1, num2):
  print(num1, '*', num2, '=', num1*num2)
def div(num1,num2):
  print(num1, '/', num2, '=', num1/num2)

#--------------------------------------------(실행)
num1 = int(input('첫번째 수 : '))
num2 = int(input('두번째 수 : '))
op = input('어떤계산을 원하시나요?(+,-,*,/) : ')
if op == '+':
  add(num1,num2)
elif op =='-':
  sub(num1,num2)
elif op=='*':
  mul(num1,num2)
elif op == '/':
  div(num1,num2)
```

```python
class calcul:
  def setup(self, a,b):   #클래스내에 있는 함수는 메서드
    self.n1 = a
    self.n2 = b
  def add(self):
    result = self.n1 + self.n2
    return result
```

```python
a = calcul()   #객체 생성 a
a.setup(2,8)    #멤버변수 정의

b = calcul()   #객체 생성 b
b.setup(32,80) #멤버변수 정의

print(a.n1)
print(b.n1)

print(a.add()) #덧셈 메서드(함수) 실행
print(b.add()) #덧셈 메서드(함수) 실행
```
